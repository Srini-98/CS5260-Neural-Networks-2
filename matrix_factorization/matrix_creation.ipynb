```python
import pandas as pd
import numpy as np
import pickle as pkl
import torch
```

```python
dataset_folder = "../dataset/"
word2vec_emb_file = dataset_folder + "Word2VecAverageArray.npy"
#word2vec_emb_file = dataset_folder + "sentenceBert.npy"
cnn_file = "../googleNet_rep.npy"
```

```python
df = pd.read_csv("../dataset/books_with_genres.csv")
df.head()
```

```python
word2vec_emb = np.load(word2vec_emb_file)
word2vec_emb.shape
```

```python
cnn_emb = np.load(cnn_file, allow_pickle=True)
```

```python
cnn_emb.shape
```

```python
df[[f"word2vec_emb{i}" for i in range(0, word2vec_emb.shape[1])]] = word2vec_emb
df.head()
```

```python
cnn_emb_dict = {}
for c in cnn_emb:
    cnn_emb_dict[list(c.keys())[0]] = c[list(c.keys())[0]]
```

```python
df["cnn_emb"] = df["book_id"].astype('str').apply(lambda x: cnn_emb_dict[x])
df.head()
```

```python
word2vec_tensor = torch.Tensor(df[[f"word2vec_emb{i}" for i in range(0, word2vec_emb.shape[1])]].values)
word2vec_tensor.size()
```

```python
x = np.zeros((5000, 1000))
for i in range(len(df["cnn_emb"])):
    x[i] = df["cnn_emb"][i].numpy()
```

```python
cnn_emb_tensor = torch.Tensor(x)
cnn_emb_tensor.size()
```
