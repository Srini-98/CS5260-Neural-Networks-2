```python
import torchvision.models as models
import torch
from torchvision import datasets, transforms
from PIL import Image
import os
import numpy as np
```

```python
data_dir = 'dataset/images/images/'
input_size = 224
```

```python
data_transforms = transforms.Compose([
        transforms.RandomResizedCrop(input_size),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

image_files = os.listdir(data_dir)
# image_datasets = datasets.ImageFolder(data_dir, data_transforms)
# dataloaders = torch.utils.data.DataLoader(image_datasets, batch_size=batch_size, shuffle=True, num_workers=4)
```

```python
googlenet = models.googlenet(pretrained=True)
googlenet.eval()
```

```python
image_rep = []
error_files = []
count = 0
for image_file in image_files:
    try:
        img = Image.open(data_dir + image_file)
        img = img.convert('RGB')
        img_t = data_transforms(img)
        batch_t = torch.unsqueeze(img_t, 0)
        model_output = {} 
        model_output[image_file] = googlenet(batch_t).detach()
        image_rep.append(model_output) 
    except:
        error_files.append(image_file)
    count += 1
    if count % 100 == 0:
        print(count)
```
