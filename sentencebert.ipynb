```python
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
df = pd.read_csv("./books_with_genres.csv")
```

```python
description_corpus = df['description']
sentences = []

description_corpus[0]
val = []
```

```python
for i in description_corpus:
    text = i
    text = re.sub(r'\[[0-9]*\]' , ' ' , text)
    text = re.sub(r'\s+' , ' ' , text)
    text = text.lower()
    text = re.sub(r'\d' , ' ' , text)
    text = re.sub(r'\s+' , ' ' , text)
    val.append(text)
```

```python
import numpy as np
arr = np.zeros((768,))
val = np.zeros((5000 , 768))

for sentence , j in zip(sentences , range(0 , 5000)):
  val[j] = sbert_model.encode(sentence)
```

```python
np.save("./sentenceBert.npy" , val)
```
