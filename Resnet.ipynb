```python
import torch
from torchvision import datasets , models , transforms
from PIL import Image
import matplotlib.pyplot as plt
import os
import torch.nn as nn
import torch.nn.functional as f
from torch.utils.data import Dataset
import pandas as pd

from torch.utils.data import Dataset
import glob
import os
```

```python
class Identity(nn.Module):
    def __init__(self):
        super(Identity , self).__init__()
    def forward(self , x):
        return x
```

```python
preprocess = transforms.Compose([
            transforms.Resize([224 , 224]),
            transforms.ToTensor(),
        ])
```

```python
model = models.resnet18(pretrained= True)

path = "C:/Users/91984/OneDrive/Desktop/NUS/courses/Semester3/NN-2/project/CS5260-Neural-Networks-2/dataset/images/"

image_lis = os.listdir(path)
```

```python
output = {}
for i in image_lis:
    print(i)
    im = Image.open(path + i)
    im_rgb = im.convert("RGB")
    input_tensor = preprocess(im_rgb)
    input_batch = input_tensor.unsqueeze(0)
    
    if torch.cuda.is_available():
        input_batch = input_batch.to('cuda')
        model.to('cuda')
    with torch.no_grad():
        output[i] = model(input_batch)
```

```python
import pickle
with open('C:/Users/91984/OneDrive/Desktop/NUS/courses/Semester3/NN-2/project/CS5260-Neural-Networks-2/ResNet18.pickle', 'wb') as handle:
    pickle.dump(output,handle)
```
