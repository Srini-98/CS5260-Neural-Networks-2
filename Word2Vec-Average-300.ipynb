```python
import pandas as pd
from gensim.models import Word2Vec
import gensim
import gensim.downloader as gensim_api
from nltk.corpus import stopwords
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
model = gensim_api.load("word2vec-google-news-300")
```

```python
model.get_vector("similar").shape
```

```python
df = pd.read_csv("./dataset/books_with_genres.csv")
```

```python
print(df.columns)
```

```python
description_corpus = df['description']
```

```python
len(description_corpus)
sentences = []

description_corpus[0]
val = []
```

```python
#model = gensim.models.Word2Vec.load_word2vec_format('./model/GoogleNews-vectors-negative300.bin', binary=True)  
```

```python
for i in description_corpus:
    text = i
    text = re.sub(r'\[[0-9]*\]' , ' ' , text)
    text = re.sub(r'\s+' , ' ' , text)
    text = text.lower()
    text = re.sub(r'\d' , ' ' , text)
    text = re.sub(r'\s+' , ' ' , text)
    val.append(text)
```

```python
sentences = [nltk.word_tokenize(st) for st in val]
```

```python
len(sentences)
```

```python
for i in range(0 , len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
sentences[0]
```

```python
import numpy as np
arr = np.zeros((300,))
val = np.zeros((5000 , 300))

for sentence , j in zip(sentences , range(0 , 5000)):
    print("j is" , j)
    try:
        arr = np.zeros((300,))
        for word in sentence:
            arr = arr + model.get_vector(word)
        arr /= len(sentence)
        val[j] = arr
    except:
        continue
    
    
```

```python
np.save("./dataset/word2vec_300_array.npy" , val)

val.shape
```
